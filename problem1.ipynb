```python
import math

def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    # Differences in coordinates
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    
    return c
```

```python
from dataclasses import dataclass

@dataclass
class AirportInfo:
    index: int
    airport_code: str
    latitude_deg:float
    longitude_deg:float

all_airports = []
for index, row in airports.iterrows():
    new_airport_info = AirportInfo(
        index=index,
        airport_code=row["airport_code"],
        latitude_deg=float(row["latitude_deg"]),
        longitude_deg=float(row["longitude_deg"]),        
    )
    all_airports.append(new_airport_info)

for airport in all_airports:
    print(airport)
```

```python
num_airports = len(all_airports)
```

```python
d = []
for i in range(num_airports):
    latitude_deg_i = all_airports[i].latitude_deg
    longitude_deg_i = all_airports[i].longitude_deg
    distance_i = []
    for j in range(num_airports):
        latitude_deg_j = all_airports[j].latitude_deg
        longitude_deg_j = all_airports[j].longitude_deg
        distance_i_j = haversine(latitude_deg_i, longitude_deg_i, latitude_deg_j, longitude_deg_j)
        distance_i.append(distance_i_j)
    d.append(distance_i)
    print(i, [f"{num:.2f}" for num in distance_i])
```

```python
from gurobipy import Model, GRB
```

```python
model = Model("Optimizing_Shipment")
```

```python
# decision variable for every pair of indices
x = {(i, j): model.addVar(vtype=GRB.BINARY, name=f"x_{i}_{j}")
                         for i in range(num_airports)
                         for j in range(num_airports)}

for i in range(num_airports):
    model.addConstr(x[i, i] == 0, name=f"no_self_loop_{i}")
```

```python
# Set the objective function, minimize d_i_j * x_i_j
model.setObjective(
    sum(d[i][j] * x[i, j] for i in range(num_airports) for j in range(num_airports)),
    GRB.MINIMIZE 
)
```

```python
# Exactly one trip leaving each airport, except LAX
for i in range(num_airports):
    if i == 1:
        continue
    model.addConstr(sum(x[i, j] for j in range(num_airports) if i != j) == 1, name=f"leaving_constraint_{i}")

# Exactly one trip entering each airport, except JFK
for i in range(num_airports):
    if i == 0:
        continue
    model.addConstr(sum(x[j, i] for j in range(num_airports) if i != j) == 1, name=f"entering_constraint_{i}")
```

```python
# MTZ constraints
u = {i: model.addVar(vtype=GRB.INTEGER, lb=1, ub=num_airports, name=f"u_{i}") for i in range(num_airports)}

model.addConstr(u[0]==1, name="JFK is the starting point")
model.addConstr(u[1]==num_airports, name="LAX is the destination")

# Add MTZ subtour elimination constraints for all i, j in N, i != j
# Note, this place can go wrong
for i in range(num_airports):
    for j in range(num_airports):
        if i != j and i != 0 and j != 0:
            model.addConstr(u[i] + 1 <= u[j] + (num_airports - 1) * (1 - x[i, j]), name=f"mtz_constraint_{i}_{j}")
```

```python
# Update model to integrate new variables
model.update()
```

```python
# Print all variables (before optimization)
print("Variables in the model:")
for v in model.getVars():
    print(v.varName)
```

```python
# Set parameters for near-optimal solution
model.setParam("MIPGap", 0.1)

# Optimize the model
model.optimize()
```

```python
print("Objective Function:", model.getObjective())
```

```python
# Print constraints
print("\nConstraints in the model:")
for constr in model.getConstrs():
    constr_name = constr.ConstrName
    lhs = model.getRow(constr)  # Get left-hand side
    rhs = constr.RHS            # Get right-hand side
    sense = constr.Sense        # Get the sense of the constraint

    # Print constraint in human-readable form
    if sense == '<':
        sense_symbol = "<="
    elif sense == '>':
        sense_symbol = ">="
    else:
        sense_symbol = "=="
        
    print(f"{constr_name}: {lhs} {sense_symbol} {rhs}")
```

```python
# Print all variables (after optimization)
print("\nVariables and their values after optimization:")
if model.status == GRB.OPTIMAL:
    for v in model.getVars():
        print(f"{v.varName}: {v.x}")
else:
    print("No optimal solution found.")
```
